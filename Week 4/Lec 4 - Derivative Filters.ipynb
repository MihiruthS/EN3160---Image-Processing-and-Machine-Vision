```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('images/sigiriya.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
# Prewitt operator
kx = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
ky = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
```

```python
imavgx = cv.filter2D(img, cv.CV_32F, kx)
imavgy = cv.filter2D(img, cv.CV_32F, ky)
```

```python
#Display the 2 images in subplots labeled as 'Prewitt X' and 'Prewitt Y'
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(imavgx, cmap='gray')
plt.title('Prewitt X')
plt.subplot(1, 2, 2)
plt.imshow(imavgy, cmap='gray')
plt.title('Prewitt Y')
plt.show()
```

```python
#sobel operator
sx = np.array([[-1, -2, -1], [0, 0, 0], [1,2,1]])
sy = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])

imgsx = cv.filter2D(img, cv.CV_32F, sx)
imgsy = cv.filter2D(img, cv.CV_32F, sy)
```

```python
#display the 2 images in subplots labeled as 'Sobel X' and 'Sobel Y'
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(imgsx, cmap='gray')
plt.title('Sobel X')
plt.subplot(1, 2, 2)
plt.imshow(imgsy, cmap='gray')
plt.title('Sobel Y')
plt.show()
```
