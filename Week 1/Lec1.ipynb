```python
# Increasing the brightness
import cmath
import cv2 as cv
import numpy as np
from matplotlib import cm
import matplotlib.pyplot as plt

# Load the image
im1 = cv.imread('/content/drive/MyDrive/Image Processing EN3160/emma.jpg', cv.IMREAD_GRAYSCALE)
assert im1 is not None

# Increase the brightness
im2 = cv.convertScaleAbs(im1, alpha=1, beta=100)

# Create subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Display the original image
ax[0].imshow(cv.cvtColor(im1, cv.COLOR_BGR2RGB), cmap = cm.gray, vmin = 0, vmax = 255)
ax[0].axis('off')
ax[0].set_title('Original Image')

# Display the brightened image
ax[1].imshow(cv.cvtColor(im2, cv.COLOR_BGR2RGB), cmap = cm.gray, vmin = 0, vmax = 255)
ax[1].axis('off')
ax[1].set_title('Brightened Image')

# Show the plot
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

f = cv.imread('/content/drive/MyDrive/Image Processing EN3160/emma.jpg', cv.IMREAD_GRAYSCALE)
assert f is not None

M,N = f.shape

h= cv.calcHist([f],[0],None,[256],[0,256])
cdf = np.cumsum(h)
L = 256
t = np.uint8((L-1)*cdf/ M*N)
g = t[f]


#Display the original image and quilized image side by side
#Label the original image as Original and Eualized one as Equalized
#Turn the axis off in both

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(cv.cvtColor(f, cv.COLOR_BGR2RGB), cmap = cm.gray, vmin = 0, vmax = 255)
ax[0].axis('off')
ax[0].set_title('Original Image')

ax[1].imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB), cmap = cm.gray, vmin = 0, vmax = 255)
ax[1].axis('off')
ax[1].set_title('Equalized Image')
```
