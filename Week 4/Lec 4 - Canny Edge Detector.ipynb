```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('images\keira_gray.jpg', cv.IMREAD_GRAYSCALE)

# Perform Canny edge detection
edges = cv.Canny(img, threshold1=100, threshold2=200)

# Display the original image and the edges in subplots
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edge Image')
plt.show()
```

```python
# Perform Canny edge detection using only an upper threshold
upper_threshold = 150
lower_threshold = 1  # Set lower threshold very low
edges = cv.Canny(img, lower_threshold, upper_threshold)

# Display the original image and the edges in subplots
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edge Image')
plt.show()
```

```python
# Perform Canny edge detection using only an upper threshold
upper_threshold = 150
lower_threshold = 140
edges = cv.Canny(img, lower_threshold, upper_threshold)

# Display the original image and the edges in subplots
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edge Image')
plt.show()
```

```python
# Perform Canny edge detection using only an upper threshold
upper_threshold = 400 # Set upper threshold very high
lower_threshold = 190
edges = cv.Canny(img, lower_threshold, upper_threshold)

# Display the original image and the edges in subplots
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edge Image')
plt.show()
```

### This is called Hysteresis Thresholding

Use the High threshold to start edge curves and the low threshold to continue them
